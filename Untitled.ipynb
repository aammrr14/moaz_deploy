# building the model

```python
ohe = OneHotEncoder(use_cat_names=True)
x=ohe.fit_transform(df["region"])
df=pd.concat([df,x], axis=1)
df.drop(columns=["region"], inplace=True)
```

```python
labelencoder = LabelEncoder()

df['sex'] = labelencoder.fit_transform(df['sex'])
df['smoker'] = labelencoder.fit_transform(df['smoker'])
```

```python
df
```

```python
target = "charges"
y = df[target]
X = df.drop(columns=target)
```

```python
'''
sc = StandardScaler()
sc.fit(X[["age","sex","bmi"]])
df1 = sc.transform(X[["age","sex","bmi"]])
df1=pd.DataFrame(df1)
col=list(pd.Series(["age","sex","bmi"]))
df1.set_axis(col, axis='columns', inplace=True)
df1.head()
X[["age","sex","bmi"]]=df1[["age","sex","bmi"]]
'''
```

```python
TI_train,TI_test,S_train,S_test = train_test_split (X,y,test_size = 0.2,random_state=38)
```

# Linear regression

```python
model_linear= LinearRegression()
model_linear.fit(TI_train,S_train)
linear_predict=model_linear.predict(TI_train)
print("Training MAE:", mean_absolute_error(S_train,linear_predict))
print("Test MAE:", mean_absolute_error(S_test,model_linear.predict(TI_test)))
```

```python
model_linear.score(TI_test,S_test)
```

# Ridge

```python
model_Ridge= Ridge()
model_Ridge.fit(TI_train,S_train)
Ridge_predict=model_Ridge.predict(TI_train)
print("Training MAE:", mean_absolute_error(S_train,Ridge_predict))
print("Test MAE:", mean_absolute_error(S_test,model_Ridge.predict(TI_test)))
```

```python
model_Ridge.score(TI_test,S_test)
```

# Random forest

```python
model_RandomForest= RandomForestRegressor()
model_RandomForest.fit(TI_train,S_train)
RandomForest_predict=model_RandomForest.predict(TI_train)
print("Training MAE:", mean_absolute_error(S_train,RandomForest_predict))
print("Test MAE:", mean_absolute_error(S_test,model_RandomForest.predict(TI_test)))
```

```python
model_RandomForest.score(TI_test,S_test)
```

# Gboost

```python
model_GradientBoosting= GradientBoostingRegressor(n_estimators=100,max_depth=3)
model_GradientBoosting.fit(TI_train,S_train)
GradientBoosting_predict=model_GradientBoosting.predict(TI_train)
print("Training MAE:", mean_absolute_error(S_train,GradientBoosting_predict))
print("Test MAE:", mean_absolute_error(S_test,model_GradientBoosting.predict(TI_test)))
```

```python
model_GradientBoosting.score(TI_test,S_test)
```
